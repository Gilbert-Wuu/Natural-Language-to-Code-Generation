```python
import json
import glob

# Find InstructHumanEval results files
results_files = glob.glob('/tmp/*humaneval*.results.json')

# If not found, look for the evaluation results JSON
if not results_files:
    # Check if there's a main results file
    eval_results_files = glob.glob('*humaneval*.json')
    print(f"Looking for evaluation results files: {eval_results_files}")

    if eval_results_files:
        for file in eval_results_files:
            print(f"\n{'='*60}")
            print(f"File: {file}")
            print('='*60)
            with open(file, 'r') as f:
                results = json.load(f)
            print(json.dumps(results, indent=2))
else:
    print(f"Found {len(results_files)} InstructHumanEval result files\n")

    for result_file in results_files[:3]:
        print(f"{'='*60}")
        print(f"File: {result_file}")
        print('='*60)
        with open(result_file, 'r') as f:
            result = json.load(f)

        # Show first result
        if result.get('results') and len(result['results']) > 0:
            first_result = result['results'][0]
            print(f"Status: {first_result.get('status')}")
            print(f"Exit code: {first_result.get('exit_code')}")
            if first_result.get('stderr'):
                print(f"Stderr: {first_result['stderr'][:500]}")
            if first_result.get('stdout'):
                print(f"Stdout: {first_result['stdout'][:500]}")
        print()
```
