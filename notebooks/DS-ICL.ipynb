# Prefix Template

```python
# --prefix "$PREFIX"

PREFIX_0_SHOT = '''
You are an expert Python programmer.
Given a problem description, write a correct Python function that satisfies the requirements.
Return only executable Python code and nothing else.
'''

PREFIX_1_SHOT = """
Here are some examples:
### Example 1
def factorial(n):
    if n == 0 or n == 1: return 1
    return n * factorial(n - 1)
"""

PREFIX_3_SHOT = """
Here are some examples:

### Example 1
def longest_string(str1, str2):
    if len(str1) > len(str2):
        return str1
    else:
        return str2


### Example 2
def decimal_to_binary(num):
    binary = []

    while num > 0:
        binary.append(num%2)
        num //= 2
        binary.reverse()
    return binary]


### Example 3
def reverse_list(lst):
    i, j = 0, len(lst) - 1
    while i < j:
        lst[i], lst[j] = lst[j], lst[i]
        i += 1
        j -= 1
    return lst
"""


PREFIX_5_SHOT = """
Here are some examples:

### Example 1
def longest_string(str1, str2):
    if len(str1) > len(str2):
        return str1
    else:
        return str2


### Example 2
def decimal_to_binary(num):
    binary = []

    while num > 0:
        binary.append(num%2)
        num //= 2
        binary.reverse()
    return binary]


### Example 3
def reverse_list(lst):
    i, j = 0, len(lst) - 1
    while i < j:
        lst[i], lst[j] = lst[j], lst[i]
        i += 1
        j -= 1
    return lst


### Example 4
def solve(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    fib = [0, 1]
    for i in range(2, n):
        fib.append(fib[-1] + fib[-2])
    return fib


### Example 5
def fibonacci(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]
    seq = [0, 1]
    for i in range(2, n):
        seq.append(seq[-1] + seq[-2])
    return seq

"""


```
